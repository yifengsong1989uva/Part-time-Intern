```python
mydata=pd.read_csv("Site Activity by Day of Week during last 10 weeks.csv")
mydata
```

```python
day_dict={0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"}
```

```python
mydata.sort_values(["Day of Week"])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.bar(np.array(mydata['Day of Week']),np.array(mydata['Sessions']),align="center",color="green")
plt.xticks(np.array(mydata['Day of Week']),[day_dict[i] for i in np.array(mydata['Day of Week'])],rotation=90,fontsize=12)
plt.ylabel("Number of Sessions",fontsize=12)
plt.title("Number of Visits by Day of Week (last 10 weeks)",fontsize=18)
```

```python
plt.bar(np.array(mydata['Day of Week']),np.array(mydata.iloc[:,2]),align="center",color="green")
plt.xticks(np.array(mydata['Day of Week']),[day_dict[i] for i in np.array(mydata['Day of Week'])],rotation=90,fontsize=12)
plt.ylabel("Bounce Rate",fontsize=12)
plt.ylim(min(np.array(mydata.iloc[:,2]))-10,max(np.array(mydata.iloc[:,2]))+10)
plt.title("Bounce Rate by Day of Week (last 10 weeks)",fontsize=18)
```

```python
plt.bar(np.array(mydata['Day of Week']),np.array(mydata.iloc[:,3]),align="center",color="green")
plt.xticks(np.array(mydata['Day of Week']),[day_dict[i] for i in np.array(mydata['Day of Week'])],rotation=90,fontsize=12)
plt.ylabel("Avg. Session Duration (sec.)",fontsize=12)
plt.title("Avg. Session Duration by Day of Week (last 10 weeks)",fontsize=18)
```

```python
plt.bar(np.array(mydata['Day of Week']),np.array(mydata.iloc[:,4]),align="center",color="green")
plt.xticks(np.array(mydata['Day of Week']),[day_dict[i] for i in np.array(mydata['Day of Week'])],rotation=90,fontsize=12)
plt.ylabel("Avg. Number of Pages per Session",fontsize=12)
plt.title("Avg. Number of Pages per Session by Day of Week (last 10 weeks)",fontsize=12)
```

```python
plt.bar(np.array(mydata['Day of Week']),np.array(mydata.iloc[:,5]),align="center",color="green")
plt.xticks(np.array(mydata['Day of Week']),[day_dict[i] for i in np.array(mydata['Day of Week'])],rotation=90,fontsize=12)
plt.ylabel("Avg. Time on Page (sec.)",fontsize=12)
plt.title("Avg. Time on Page by Day of Week (last 10 weeks)",fontsize=18)
```
