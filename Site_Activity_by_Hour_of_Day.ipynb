```python
mydata=pd.read_csv("Site Activity by Hour of Day during last 10 weeks.csv")
#mydata=mydata.iloc[:-1,]
mydata
```

```python
day_dict={}
for i in mydata.Hour:
    day_dict[i]=str(i)+"-"+str(i+1)
day_dict
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,9))
plt.bar(np.array(mydata['Hour']),np.array(mydata['Sessions']),align="center")
plt.xticks(np.array(mydata['Hour']),[day_dict[i] for i in np.array(mydata['Hour'])],rotation=90,fontsize=15)
plt.xlim(-1,24)
plt.xlabel("Hour of Day", fontsize=15)
plt.ylabel("Number of Sessions",fontsize=15)
plt.title("Number of Visits by Hour of Day (last 10 weeks)",fontsize=18)
```

```python
plt.figure(figsize=(12,9))
plt.bar(np.array(mydata['Hour']),np.array(mydata.iloc[:,2]),align="center")
plt.xticks(np.array(mydata['Hour']),[day_dict[i] for i in np.array(mydata['Hour'])],rotation=90,fontsize=15)
plt.xlim(-1,24)
plt.xlabel("Hour of Day", fontsize=15)
plt.ylabel("Bounce Rate",fontsize=15)
plt.ylim(min(np.array(mydata.iloc[:,2]))-10, max(np.array(mydata.iloc[:,2]))+10)
plt.title("Bounce Rate by Hour of Day (last 10 weeks)",fontsize=18)
```
